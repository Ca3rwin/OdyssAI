```python
import spacy
num_parts = 9                                                                   

# Crea la funzione per dividere il testo in frasi
def split_in_sentences(text):
    doc = nlp(text)
    return [str(sent).strip() for sent in doc.sents]

# Crea l'oggetto nlp per l'elaborazione di linguaggio italiano
nlp = spacy.load('it_core_news_lg')
nlp.add_pipe('sentencizer')
nlp.max_length = 6000000

# Itera l'operazione per tutte un determinato numero di parti del dataset
for i in range(1, num_parts):
  # Apre, legge e richiude il file di testing
  f = open(f'part_{i}.txt', 'r')
  text = f.read()
  f.close()

  # Scrive in append sul file contenente il dataset suddiviso in frasi
  f1 = open('complete_dataset_sentences.txt', 'a+')
  outputs = split_in_sentences(text)
  for output in outputs:
    f1.write(f'{output}\n')
    print(output)
  f1.close()
```
