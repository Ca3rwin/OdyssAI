```python
from scipy.stats import wilcoxon
from statsmodels.stats.descriptivestats import sign_test
import numpy as np
```

```python
# CONVERTS FILE STRING TO LIST
def fileStringToList(filename):
  f = open(filename, 'r')
  lines = f.readlines()
  f.close()
  line = lines[1].replace('[', '').replace(']\n', '')
  score_list = line.split(', ')

  return score_list
```
