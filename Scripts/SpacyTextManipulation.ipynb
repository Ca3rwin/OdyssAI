# Conteggio dei token per frase, calcolo e visualizzazione valori statistici

```python
"""Conta il numero di token di un determinato testo"""
def countTokensSpacy(text):
  doc = nlp(text)
  count = 0
  for token in doc:
    count += 1
  return count;
  
def groupSentencesByTokenCount (tokenList):
  mapToken = {} 
  for token in tokenList:
    if token in mapToken:
      mapToken[token] = mapToken[token] + 1
    else:
      mapToken[token] = 1
  return mapToken
```

```python
import spacy
from IPython.display import clear_output 

nlp = spacy.load('it_core_news_lg')
import numpy as np

f = open('testing.txt', 'r')
lines = f.readlines()
f.close()
print(len(lines))
print(lines[0:100])
```

```python
# lineLen contiene il numero di token di ogni frase relativamente al loro indice 
lineLen = []
for i, line in enumerate(lines):
  count = countTokensSpacy(line)
  lineLen.append(count - 1)
  if i % 1000 == 0:
    clear_output()
    print(i)
  
print(lineLen)
```

```python
for line in lineLen[:10]:
  print(line, type(line))
```

```python
mapToken = groupSentencesByTokenCount(lineLen)
```

```python
try:
  tokenMin = int(input('Seleziona il numero minimo di token per frase: '))
  f = open(f'testing_min_tokens_{tokenMin}.txt', 'w+')
  i, j = 0, 0
  for line in lines:
    if(lineLen[i] >= tokenMin):
      j += 1
      f.write(lines[i])
    i += 1

  print(f'Numero righe con almeno {tokenMin} token: {j}')
except:
  print('Inserisci un intero!')
```

# Fix delle virgolette e randomizzazione del dataset dopo averlo suddiviso per frasi

```python
"""Conta il numero di virgolette in una frase. Se è pari,
il numero di virgolette è ritenuto esatto, se è dispari, è ritenuto errato"""
def countQuotes(x, count):
  if x.count('"') % 2 != 0 and x.count('"') > 1:
    count[0] += 1
    # print(x)
    pass
  elif x.count('"') == 1:
    if len(x) - (x.find('"') + 1) >= 2:
      # print(x)
      count[1] += 1
    elif x.find('"') == 0:
      # x = x + '"'
      pass
    elif len(x) - (x.find('"') + 1) <= 1:
      count[2] += 1
    # print(x)
  return count

"""Se il numero di virgolette è 1, e le virgolette sono chiuse esattamente
alla fine della frase, si presume che siano dovute iniziare ad inizio frase,
quindi si aggiungono in quella posizione"""
def fixQuotes(x):
  x = x.replace('\n', '')
  if x.count('"') % 2 != 0 and x.count('"') > 1:
    # print(x)
    pass
  elif x.count('"') == 1:
    if len(x) - (x.find('"') + 1) <= 1:
      x = '"' + x
    elif x.find('"') == 0:
      x = x + '"'
    print(x)

  return x
```

```python
f = open('dataset_randomized.txt', 'r')
lines = f.readlines()
f.close()
f = open('dataset_randomized_fixed_quotes.txt', 'w+')
for line in lines:
  f.write(fixQuotes(line) + '\n')
f.close()
```

```python
"""Prende una frase casuale da una lista contenente
tutte le frasi del dataset e la scrive su un nuovo file.
Successivamente la frase viene eliminata dalla lista per
evitare che venga estratta molteplici volte"""
def randomizeDataset(datasetLines):
  print(len(datasetLines))

  f=open('dataset_randomized.txt', 'w+')

  while len(datasetLines) != 0:
    index = randint(0, len(datasetLines) - 1)
    # print(lines[index])
    # print(len(lines))
    f.write(datasetLines[index])
    datasetLines.remove(datasetLines[index])

  f.close()
  print("END")
```

```python
from random import seed
from random import randint
# seed random number generator
seed(1)

f=open('complete_dataset_sentences_full.txt', 'r')
lines = f.readlines()
f.close()

randomizeDataset(lines)
```
